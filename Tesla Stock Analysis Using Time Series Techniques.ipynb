## Simple Moving Average

```python
#simple moving average'
df_tesla['Open'].plot(figsize=(15,6))
```

```python
df_tesla['Open: 10 days rolling']= df_tesla['Open'].rolling(window=10, min_periods=1).mean()
```

```python
df_tesla[['Open','Open: 10 days rolling']].plot(xlim=['2020-01-01','2024-04-30'],figsize=(15,6))
```

```python
df_tesla['Open: 30 days rolling']= df_tesla['Open'].rolling(window=30, min_periods=1).mean()
df_tesla['Open: 50 days rolling']= df_tesla['Open'].rolling(window=50, min_periods=1).mean()
```

```python
df_tesla[['Open','Open: 10 days rolling','Open: 30 days rolling','Open: 50 days rolling']].plot(xlim=['2020-01-01','2024-04-30'],figsize=(15,6))
```

## Cumulative Moving Average 

```python
## Expanding
# cummulative moving average
df_tesla['Open'].expanding().mean().plot(figsize=(10,6))
```

## Exponential Moving Average (EMA):

```python
# EMA Exponential Moving Average
# EMA tesla share
#lets smoothing factor -0.1
df_tesla['EMA_0.1']= df_tesla['Open'].ewm(alpha=0.1,adjust=False).mean()
```

```python
df_tesla[['Open','EMA_0.1']].plot(xlim=['2020-01-01','2024-04-30'],figsize=(15,6))
```

```python
df_tesla['EMA_0.3']= df_tesla['Open'].ewm(alpha=0.3,adjust=False).mean()
```

```python
df_tesla[['Open','EMA_0.1','EMA_0.3']].plot(xlim=['2020-01-01','2024-04-30'],figsize=(15,6))
```

## Exponential Weighted Moving Average(EWMA)

```python
# EWMA Exponential Weighted Moving Average

#lets smoothing factor -0.1
df_tesla['EMA_5 days']= df_tesla['Open'].ewm(span=5).mean()
```

```python
df_tesla[['Open','EMA_5 days']].plot(xlim=['2020-01-01','2024-04-30'],figsize=(15,6))
```
